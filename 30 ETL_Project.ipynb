```python
import pandas as pd
```

```python
pd.options.display.max_columns = 100
```

```python
df1 = pd.read_csv("C:\\Users\\User\\NBA_Full_Draft_1947-2018.csv")
df1
```

```python
df2 = pd.read_csv("C:\\Users\\User\\basketball_player_allstar.csv")
df2
```

```python
df_rename = df2.rename(columns={"player_id":"Player_ID", "last_name":"Last_Name", "first_name":"First_Name", "season_id":"Season_ID", "conference":"Conference", "league_id":"League_ID", "games_played":"Games_Played", "minutes":"Minutes", "points":"Points", "o_rebounds":"O_Rebounds", "d_rebounds":"D_Rebounds", "rebounds":"Rebounds", "assists":"Assists", "steals":"Steals", "blocks":"Blocks", "turnovers":"Turnovers", "personal_fouls":"Personal_Fouls", "fg_attempted":"FG_Attempted", "fg_made":"FG_Made", "ft_attempted":"FT_Attempted", "ft_made":"FT_Made", "three_attempted":"Three_Attempted", "three_made":"Three_Made"})
df_rename
```

```python
df_rename['Player'] = df_rename['First_Name'] +' '+ df_rename['Last_Name']
df_rename
```

```python
df_drop = df_rename.drop(['Last_Name', 'First_Name'], axis=1)
df_drop
```

```python
df_drop['Allstars'] = 'True'
df_drop
```

```python
df_join = pd.merge(df1, df_drop, on="Player", how="outer")
df_join
```

```python
df_join['Allstars'] = df_join['Allstars'].fillna(False)
df_join
```

```python
df_nonull = df_join.fillna(0)
df_nonull
```

```python
df_int = df_nonull.astype({'Unnamed: 0':'int', 'Play_Yrs':'int', 'Games':'int', 'Year':'int', 'Pick':'int', 'Season_ID':'int', 'Games_Played':'int', 'Minutes':'int', 'Points':'int', 'O_Rebounds':'int', 'D_Rebounds':'int', 'Rebounds':'int', 'Assists':'int', 'Steals':'int', 'Blocks':'int', 'Turnovers':'int', 'Personal_Fouls':'int', 'FG_Attempted':'int', 'FG_Made':'int', 'FT_Attempted':'int', 'FT_Made':'int', 'Three_Attempted':'int', 'Three_Made':'int'})
df_int
```

```python
df_drop = df_int.drop(['Unnamed: 0'], axis=1)
df_drop
```

```python
df_final = df_drop[['Team','Player','Allstars','College','Play_Yrs','Games','Year','Pick','Player_ID','Season_ID','Conference','League_ID','Games_Played','Minutes','Points','O_Rebounds','D_Rebounds','Rebounds','Assists','Steals','Blocks','Turnovers','Personal_Fouls','FG_Attempted','FG_Made','FT_Attempted','FT_Made','Three_Attempted','Three_Made']]
df_final
```

```python
df_csv = df_final.to_csv(r'C:\\Users\\User\\ETL_Final.csv', index = None, header=True)
df_csv
```
